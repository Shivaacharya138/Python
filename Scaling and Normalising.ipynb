# Scaling

```python
import pandas as pd
import numpy as np
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# set seed for reproducibility
np.random.seed(0)
```

```python
# generate 1000 data points randomly drawn from an exponential distribution
Original_data = np.random.exponential(size=1000)
```

```python
from mlxtend.preprocessing import minmax_scaling
# mix-max scale the data between 0 and 1
Scaled_data = minmax_scaling(Original_data, columns=[0])
```

```python
# plot both together to compare
```

```python
fig, axes = plt.subplots(1,2, figsize=(20,5))

sns.histplot(Original_data, ax=axes[0], kde=True, legend=False)
axes[0].set_title('Original Data')

sns.histplot(Scaled_data, ax=axes[1], kde=True, legend=False)
axes[1].set_title('Scaled Data')
```

Notice that the shape of the data doesn't change, but that instead of ranging from 0 to 10, it now ranges from 0 to 1.

# Normalization

```python
# normalize the exponential data with boxcox
```

```python
# for Box-Cox Transformation
from scipy import stats
```

```python
Normalized_data = stats.boxcox(Original_data)
```

```python
fig, axes = plt.subplots(1,2, figsize=(20,5))

plt.subplot(1,2,1)
sns.histplot(Original_data, kde=True, legend=False)
axes[0].set_title("Original Data")

plt.subplot(1,2,2)
sns.histplot(Normalized_data, kde=True, legend=False)
axes[1].set_title('Normalized Data')
```

Notice that the shape of our data has changed. Before normalizing it was almost L-shaped. But after normalizing it looks more like the outline of a bell (hence "bell curve").
